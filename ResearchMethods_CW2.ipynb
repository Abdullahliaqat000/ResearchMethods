# Data Anomaly in the EXCEL SHEET
After creating my visuaization diagram I was analyzing it for the possible trends in the data set and I identified a unique observation that people belonging to the **meat group 50-99** had lesser impact on the environment than those belonging to **meat group less than 50** which didn't seemed right and after that I have gone through the excel sheet identified the anomlay which is as follows:

image.png

The grouping column shows that **meat < 50** where as the corresponding diet group shows **meat50** which is incorrect. Similarly for the **meat 50-99** the corresponding diet group listed in the dataset is **meat**.

The first step is to correct the corresponding diet groups in the data.

Import the dataset and display the rows along with displaying the list of columns and respective datatype.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("sample_data/Results_21MAR2022_nokcaladjust (1).csv")

print("Dataset shape:", df.shape)
print("COlumn Datatypes\n",df.dtypes)
print("Column names:", df.columns)
display(df.head())
```

The grouping column is made up by combining the following three coumns:


*   Diet
*   Sex
*   Age

By splitting the grouping column we can correctly idnetify the diet group and update the data.




```python
df['diet_group'] = df['grouping'].str.split('_').str[0]
```

Display the data to confirm if the diet group has been corrected or not.

```python
display(df.head(100))
```

Identify the distinct diet groups included in the dataset.

```python
print("Distinct diet groups are:")
print(df['diet_group'].unique())

print("\nThe total number of distinct diet groups are:")
print(len(df['diet_group'].unique()))
```

Now diet groups have been correctly mapped. Identify the distinct age groups included in the dataset.

```python
print("Distinct age groups are:")
print(df['age_group'].unique())

print("\nThe total number of distinct age groups are:")
print(len(df['age_group'].unique()))
```

Combine all the columns which is impacting the environment. The combination will be used in the visualization.

```python
key_environmental_matrices = [
    'mean_ghgs', 'mean_land', 'mean_watscar', 'mean_watuse',
    'mean_eut', 'mean_bio', 'mean_acid'
]
```

Since the environmental matrices are measured in different units we need to normalize all the key environmental matrices on a same scale.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[key_environmental_matrices] = scaler.fit_transform(df[key_environmental_matrices])
```

Describe the features of the normalized environmental matrices.

```python
df[key_environmental_matrices].describe()
```

The dataset contains simulation of multiple runs. We will use the mean of all simulations. Find the mean of the all environmental matrices which will be used later on in the treemap visualization

```python
df['mean_environmental_score'] = df[key_environmental_matrices].mean(axis=1)

df[['diet_group', 'sex', 'age_group', 'mean_environmental_score'] + key_environmental_matrices].head()
```

Group the data based on the diet group, gender and age group to find the mean number of participants and mean evironmental score.

```python
df_aggregated = df.groupby(['diet_group', 'sex', 'age_group']).agg({
    'n_participants': 'mean',
    'mean_environmental_score': 'mean'
}).reset_index()
```

Sum the total number of participants to verify it matches the total or not.

```python
total_participants = df_aggregated['n_participants'].sum()
print("Total participants:", int(total_participants))
```
